```python
import tensorflow as tf
import pandas as pd
```

```python
BTC_data=pd.read_csv('/content/BTC-USD-1-DAY.csv',delimiter=',')
```

```python
BTC_data.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,6))
plt.title('Close Price History')
plt.plot(BTC_data['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.show()
```

```python
import numpy as np
tesla_close=[[x] for x in BTC_data['Close']]
tesla_close
```

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(tesla_close)
```

```python
train_len=(int(len(scaled_data)*85/100))
train_data=scaled_data[:train_len]
test_data=scaled_data[train_len:]
print(len(train_data),len(test_data))
```
